```python
import numpy as np
import folium
from folium import plugins
```

Latitude e longitude vem de http://bboxfinder.com/#-27.007138,-49.159698,-26.674459,-48.965378

Acurária do 3º decimal é de ~100m
https://gis.stackexchange.com/questions/8650/measuring-accuracy-of-latitude-and-longitude

```python
long_min = -49.159698
long_max = -48.965378
lat_min = -27.007138
lat_max = -26.674459
```

```python
N_pontos = 100
lats = np.linspace(lat_min, lat_max, N_pontos)
longs = np.linspace(long_min, long_max, N_pontos)
coords = [
    (lat, long)
    for lat in lats
    for long in longs
]
```

```python
N_samples = 300
samples = []
for _ in range(5):
    np.random.shuffle(coords)
    samples += coords[:N_samples]
```

```python
m = folium.Map(
    [np.mean([lat_min, lat_max]), np.mean([long_min, long_max])],
    zoom_start=11
)

m.add_children(
    plugins.HeatMap(samples, radius=15)
)

m
```
